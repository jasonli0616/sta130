$H_1$ (alternative hypothesis): There is some association between talking to others about hobbies and interests, and how lonely someone feels.\
The data can be formatted into a contingency table, which summarizes the frequency counts of each combination of the categories of the two variables.\

## Analysis 2

**Research question**: Are people in the LGBTQ2+ community less likely to spend time with family members?

For the independent variable, I want to use `DEMO_identity_lgbtq`, where the question is "*Do you belong to any of the following groups? (Check all that apply) - Sexual or gender minorities (e.g., LGBTQ2+)*".\
The results are binary values, someone either is or is not part of the LGBTQ2+ community.

For the dependent variable, I want to use `CONNECTION_preference_time_family_grouped`, where the question is "*How much time per week would you like to spend socializing with others from the following groups? - Family Members*".\
The results are "numerical" ordinal categorical values: No time, Less than 1 hour, 1 to 4 hours, 5 or more hours.

**Hypothesis:** People in the LGBTQ2+ community are less likely to spend time with family members.\
This is my hypothesis because peple in the LGBTQ2+ community may spend less time with family members due to non-accepting family or a variety of other reasons.

This can be visualized using two bar plots showing the number of results in the dependent variable, one for people in the LGBTQ2+ community and one for queer people not. This can be shown side-by-side or overlaid.

We can use a Fisher Exact test, which determines if there is a significant association between two categorical values in a contingency table. It is designed for 2x2 contingency tables (binary values).

$H_0$ (null hypothesis): no difference in time socializing with family members between people who do and do not identify as part of the LGBTQ2+ community.\
$H_1$ (alternative hypothesis): people in the LGBTQ2+ community are less likely to spend time socializing with family members.\
The independent variable are already binary values. We will convert the dependent variable to binary: people who do and people who do not often spend time with family.\
We can build a 2x2 contingency table using these binary values.\
We can perform a Fisher Exact test in Python (`scipy.stats`), which will give us an odds ratio (strength of association), and a p-value that we can use to either reject or fail to reject the null hypothesis.

In a Fisher Exact test, the contingency table looks like the following

|                       | Outcome 1 | Outcome 2 | Row Totals |
|-----------------------|-----------|-----------|------------|
| **Group 1**           | \( a \)   | \( b \)   | \( a + b \)|
| **Group 2**           | \( c \)   | \( d \)   | \( c + d \)|
| **Column Totals**     | \( a + c \) | \( b + d \) | \( n \)  |

Formulas:

$$P = \frac{(a + b)! (c + d)! (a + c)! (b + d)!}{a! \, b! \, c! \, d! \, n!}$$

$$\text{Odds Ratio} = \frac{a \times d}{b \times c}$$

where $a,b,c,d,n$ are the values from the table and $P$ is the p-value.
