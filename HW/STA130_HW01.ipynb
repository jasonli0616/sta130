1. Provide an example of a "use case" in which using `df.dropna()` might be preferred over using `del df['col']`

`df.dropna()` will go through the `DataFrame`, and remove all rows (or columns, if specified to do so) with missing values.  This is useful when you want to remove rows with missing data, but not remove an entire column. For example, if an entry has a missing value for age, you would want to use `df.dropna()` to remove it rather than `del df['age']` which would remove the entire age column.

2. Provide an example of "the opposite use case" in which using `del df['col']` might be preferred over using `df.dropna()`

Deleting an entire column is useful when you want to remove any data that is deemed unnecessary. This Titanic dataset has its own `PassengerId` column. Since `pandas` has its own indexing column, the `PassengerId` column may be redundant and `del df['PassengerId']` can be used to remove this column.

3. Discuss why applying `del df['col']` before `df.dropna()` when both are used together could be important

When you run `df.dropna()`, it will remove all rows with empty data. It is possible that this empty data in a row belongs to the column removed in `del df['col']`. If this is the case, a row with useful data in the other columns may be removed if `df.dropna()` is applied before `del df['col']`. Since an entire column will be removed regardless of empty values, `del df['col']` should be run first. Now, `df.dropna()` will not apply on that removed column, which could preserve some useful data.

4. Remove all missing data from one of the datasets you're considering using some combination of `del df['col']` and/or `df.dropna()` and give a justification for your approach, including a "before and after" report of the results of your approach for your dataset.

When running `df.dropna()`, a majority of the data (~77%) is dropped due to having a missing "Cabin" value. This column will not be of much use, since it is missing for so many values anyways. Let's remove this column, then drop missing values. We will keep the "Age" and "Embarked" column, since a minority of the data will be removed due to missing values.

```python
df_modified = df.copy() # make a copy, to not modify the original
del df_modified["Cabin"]
df_modified = df_modified.dropna() # remove missing values
df_modified
```

```python
df.describe()
```

```python
df_modified.describe()
```

After the modification, 179 rows of data has been removed.
The mean of survival (which is the survival rate, considering they are boolean values) has increased slightly. This could be because the age values of non-surviving passengers are unknown (some age values of survivors may be added in after-the-fact).
The mean of the "Parch" column (which ChatGPT explains is the number of parents/children this passenger has aboard) has also increased, as well as the "Fare" column.
The standard deviation of some "SibSp" (ChatGPT explains means sibling/spouse) and fare has also been changed compared to the unmodified dataset.

ChatGPT also suggested to imputation of the "Age" column, rather than removing the missing data outright. This is a new term to me, and it means replacing the missing values with the mean or median of the column (as explained by ChatGPT). This would allow us to preserve other potentially useful data that would otherwise be removed with `df.dropna()`. I will use the median rather than the mean, to be more resistant to potential outliers.

`df.dropna()` will now remove the smaller number of remaining missing values.

```python
df_imputation = df.copy() # make a copy, to not modify the original
del df_imputation["Cabin"]
df_imputation['Age'].fillna(df_imputation['Age'].median(), inplace=True) # replace empty values with median
df_imputation = df_imputation.dropna() # remove missing values
df_imputation
```

```python
df.describe()
```

```python
df_imputation.describe()
```

After the modification, we can see (count) that only 2 rows of data were lost.
The mean for each category remain mostly unchanged (small change in the decimals values).
The standard deviation for age is now smaller by 2 years, which is understandable considering that we replaced all missing values with the median (imputation).
This data is much closer to the original data than the first modified version, since it preserved a lot more data.
